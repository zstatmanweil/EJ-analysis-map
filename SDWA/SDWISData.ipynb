```python
import requests
import pandas as pd
import numpy as np
from IPython.display import display
```

```python
# Identify a few rows of data
#url = 'https://iaspub.epa.gov/enviro/efservice/SDW_VIOL_ENFORCEMENT/STATE/NJ/COMPPERBEGINDATE/>/31-DEC-11/COMPPERENDDATE/</01-JAN-17/ROWS/0:30/JSON'
# Identify the whole table
url = 'https://iaspub.epa.gov/enviro/efservice/SDW_VIOL_ENFORCEMENT/STATE/NJ/COMPPERBEGINDATE/>/31-DEC-11/COMPPERENDDATE/</01-JAN-17/JSON'
r = requests.get(url)
if r.status_code == 200:
    violations = r.json()
else:
    raise Exception("Request was not valid. Code: {code}".format('code'=r.status_code))
```

```python
headers = list(violations[0].keys())
headers
```

```python
vio_data = []
for violation in violations:
    vio_data.append(list(violation.values()))
```

```python
vio_df = pd.DataFrame(vio_data, columns = headers)
with pd.option_context('display.max_columns', None):
    display(vio_df)
```

```python
vio_df['health_based'] = np.where(
                        (np.logical_or(vio_df['VTYPE']=='MR',vio_df['VTYPE']=='OTHER')),'N','Y')
```

```python
vio_df
```

```python
pivot = pd.pivot_table(vio_df, index=['PWSID','VIOID','health_based'])
#pivot = pivot.reset_index() 
pivot
```

```python
vio_by_pws = pd.pivot_table(pivot, values=['VIOID'],index=['PWSID'],aggfunc='count')
vio_by_pws = vio_by_pws.reset_index()
vio_by_pws
```
